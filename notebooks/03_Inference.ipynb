```python
from fastai.vision import *
```

```python
path = Path('../data/training')
learn = load_learner(path)
```

```python
img = open_image(path/'assorted_drinks.JPG')

pred_class, pred_idx, outputs = learn.predict(img)
pred_class

bbox = ImageBBox.create(*img.size, [[2, 45, 360, 250]], labels=[0],
                        classes=['3.54/5 (604 Ratings) \n' + pred_class.__str__().upper()])
img.show(figsize = (8, 8), y=bbox)
```
